```python
def helper_k(object):
    keywords=[]
    counter=0
    for i in ast.literal_eval(object):
        if counter!=3:
            keywords.append(i["name"])
            counter+=1
        else:
            break
    return keywords
```

```python
movies["keywords"]=movies["keywords"].apply(helper_k)
```

```python
movies.head()
```

```python
def helper_c(object):
    L=[]
    for i in ast.literal_eval(object):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"]=movies["crew"].apply(helper_c)
```

```python
movies.head()
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["Tags"]=movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
Df=movies[["id","title","Tags"]]
```

```python
Df.head()
```

```python
Df["Tags"]=Df["Tags"].apply(lambda x:" ".join(x))
```

```python
Df["Tags"][0]
```

```python
Df["Tags"]=Df["Tags"].apply(lambda x:x.lower())
```

```python
Df.head()
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(Df["Tags"]).toarray()
```

```python
vector[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemm(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
Df["Tags"]=Df["Tags"].apply(stemm)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarly=cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarly[0])), reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=Df[Df["title"]==movie].index[0]
    distance=similarly[movie_index]
    movie_list=sorted(list(enumerate(distance)), reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(Df.iloc[i[0]].title)
```

```python
recommend("The Dark Knight")
```
